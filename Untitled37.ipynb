```python
class Persona:
  # Constructor
  def __init__(self, nombre, edad, estatura, peso):
    self.nombre = nombre # Atributo
    self.edad = edad # Atributo
    self.estatura = estatura # Atributo
    self.peso = peso # Atributo


  def imc(self):
    return self.peso / self.estatura ** 2

  def saludar(self):
    print("Hola", self.nombre, "tu IMC es", self.imc())
```

```python
persona1 = Persona("Erick", 20, 1.70, 58)
persona1.imc()
```

```python
persona1.saludar()
```

```python
persona2 = Persona("Juan", 25, 1.72, 78)
persona2.saludar()
```

```python
# Mostrar atributos
print(persona1.nombre)
print(persona1.edad)
```

```python
class Calculadora:
  def __init__(self, a, b):
    self.a = a # Atributo
    self.b = b # Atributo

  def sumar(self):
    return self.a + self.b # Metodo

  def multiplicar(self):
    return self.a * self.b # Metodo


  def dividir(self):
    try:
      return self.a / self.b # Metodo
    except ZeroDivisionError:
      print("No se puede dividir por cero")
```

```python
calcula1 = Calculadora(3, 5)
calcula1.dividir()
```

```python
calcula2 = Calculadora(2, 0)
calcula2.dividir()
```

```python
calc = Calculadora(3, 4)

print("Suma: ", calc.sumar())
print("Multiplicación: ", calc.multiplicar())
```

```python
try:
  a = float(input("Ingrese un número real: "))
  b = float(input("Ingrese un segundo número real: "))

  calc = Calculadora(a, b)

  print("División: ", calc.dividir())

except ValueError:
  print("Debes ingresar números válidos")
except ZeroDivisionError:
  print("No se puede dividir por cero")
```

```python
class Estudiante:

  """Métodos aprobar."""

  def __init__(self, nombre, nota):
    self.nombre = nombre
    self.nota = nota

  def aprobar(self):
    if self.nota >= 3.0:
      print(self.nombre, "Aprobó")
    else:
      print(self.nombre, "Reprobo")
```

```python
class CuentaBancaria:

  def __init__(self, titular, saldo):
    self.titular = titular
    self.saldo = saldo

  def depositar(self, cantidad):
    self.saldo += cantidad # Al saldo le sumamos la cantidad

  def retirar(self, cantidad):
    self.saldo -= cantidad # Al saldo le restamos la cantidad

  def mostrar_saldo(self):
    print("El saldo del titular luego del movimiento es: ", self.saldo)
```

```python
cuenta = CuentaBancaria("Ercik", 1000)
cuenta.depositar(500)
cuenta.retirar(300)
cuenta.mostrar_saldo()
```
