Lenguaje regulares

```python
import re

try:

  nombre = input("Ingrese su nombre y apallido sin espacios: ").lower()
  edad = input("Ingrese su edad: ")

  if not re.fullmatch(r"[a-zA-Z]+", nombre):
    raise ValueError("El nombre solo debe contener letras")

  if not re.fullmatch(r"\d+", edad):
    raise ValueError("la edad debe contener solo números")

  contraseña = nombre[:3]+ edad + "@2026"
  print("Contraseña generada: ")
  print(contraseña)

except ValueError as e:
  print("Error: ", e)
```

```python
import re #código de chat

try:
    usuario = input("Ingrese su usuario sin espacios: ").lower()
    telefono = input("Ingrese su número de teléfono: ")

    # Validar que el usuario solo tenga letras
    if not re.fullmatch(r"[a-zA-Z]+", usuario):
        raise ValueError("El usuario solo debe contener letras")

    # Validar que el teléfono solo tenga números
    if not re.fullmatch(r"\d+", telefono):
        raise ValueError("El teléfono solo debe contener números")

    # Generar código de acceso
    codigo = usuario[:4] + telefono[-3:] + "#2026"

    print("Código generado:")
    print(codigo)

except ValueError as e:
    print("Error:", e)
```
