```python
import io
e_capitulos = io.BytesIO(uploaded['e_capitulos (1).csv'])
e_capitulos = pd.read_csv(e_capitulos, sep = ",", encoding = "latin-1")
e_capitulos.head(4)
```

```python
tabla_cigarrillo = pd.DataFrame({
    'Consumo': e_capitulos['E_01'], #¿Ha fumado cigarrillos/tabaco alguna vez en la vida?
    'edad': e_capitulos['E_02'], # ¿Qué edad tenía usted cuando fumó cigarrillos/tabaco por primera vez?
    'fumar': e_capitulos['E_06'],   # ¿Cuántos días ha fumado cigarrillos/tabaco durante los últimos 30 días?
    'frecuencia mensual': e_capitulos['E_05'], # ¿Ha fumado cigarrillos/tabaco en los últimos 30 días?
    'Consumo_e': e_capitulos['E_10'], # ¿Ha fumado cigarrillos electrónicos o vaporizadores con nicotina alguna vez en la vida?
    'edad_electronico': e_capitulos['E_11'], # ¿Qué edad tenía usted cuando los uso por primera vez?
    'frecuencia mensual_e': e_capitulos['E_12'], # ¿Ha usado cigarrillos electrónicos o vaporizadores con nicotina en los últimos 30 días?
})
tabla_cigarrillo.head(5)
```

```python
tabla_cigarrillo['fumar'] = tabla_cigarrillo['fumar'].fillna(0)
tabla_cigarrillo.head(5)
```

```python
tabla_cigarrillo['frecuencia mensual'] = tabla_cigarrillo['frecuencia mensual'].fillna(0)
tabla_cigarrillo.head(5)
```

```python
tabla_cigarrillo['frecuencia mensual_e'] = tabla_cigarrillo['frecuencia mensual_e'].fillna(0)
tabla_cigarrillo.head(5)
```

```python
tabla_cigarrillo = tabla_cigarrillo.dropna()
tabla_cigarrillo.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Asegurar datos numéricos
tabla_cigarrillo['edad'] = pd.to_numeric(tabla_cigarrillo['edad'], errors='coerce')

# Eliminar nulos
edad = tabla_cigarrillo['edad'].dropna()

# Crear bins de 1 en 1 (año a año)
bins = range(int(edad.min()), int(edad.max()) + 1)

plt.figure()
plt.hist(edad, bins=bins)

plt.title("Distribución de edades (año a año)")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")

plt.show()
```

```python
dataf = pd.get_dummies(
    tabla_cigarrillo,
    columns=['Consumo', 'frecuencia mensual', "Consumo_e", "frecuencia mensual_e"],
    prefix=['Consumo', 'frecuencia mensual',  "Consumo_e", "frecuencia mensual_e"]
)
dataf.head(4)

```

```python
dataf['Consumo_1'] = dataf['Consumo_1'].astype(int)
dataf['frecuencia mensual_0.0'] = dataf['frecuencia mensual_0.0'].astype(int)
dataf['frecuencia mensual_1.0'] = dataf['frecuencia mensual_1.0'].astype(int)
dataf['frecuencia mensual_2.0'] = dataf['frecuencia mensual_2.0'].astype(int)
dataf['frecuencia mensual_9.0'] = dataf['frecuencia mensual_9.0'].astype(int)
dataf['Consumo_e_1'] = dataf['Consumo_e_1'].astype(int)
dataf['frecuencia mensual_e_1.0'] = dataf['frecuencia mensual_e_1.0'].astype(int)
dataf['frecuencia mensual_e_2.0'] = dataf['frecuencia mensual_e_2.0'].astype(int)
dataf.head(4)
```
