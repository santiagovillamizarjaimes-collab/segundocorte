```python
frecuencia_cigarrillo = (
    tabla_cigarrillo['fumar']
    .value_counts()
    .sort_index()
)

frecuencia_cigarrillo = frecuencia_cigarrillo[
    frecuencia_cigarrillo > 10
]

plt.figure(figsize=(10,5))
frecuencia_cigarrillo.plot(kind='bar')

plt.title("Consumo de cigarrillo (categorías frecuentes)")
plt.xlabel("Número de cigarrillos")
plt.ylabel("Frecuencia")
plt.xticks(rotation=0)

plt.show()
```

```python
dataf = pd.get_dummies(
    tabla_cigarrillo,
    columns=['Consumo', "Consumo_e", "frecuencia mensual_e"],
    prefix=['Consumo',  "Consumo_e", "frecuencia mensual_e"]
)
dataf.head(4)

```

```python
dataf['Consumo_1'] = dataf['Consumo_1'].astype(int)
dataf['Consumo_e_1'] = dataf['Consumo_e_1'].astype(int)
dataf['frecuencia mensual_e_1.0'] = dataf['frecuencia mensual_e_1.0'].astype(int)
dataf['frecuencia mensual_e_2.0'] = dataf['frecuencia mensual_e_2.0'].astype(int)
dataf.head(4)
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(dataf, test_size=0.2, random_state=42)
```

```python
X_train = train_data.drop('frecuencia mensual', axis=1) # Variables explicativas
y_train = train_data['frecuencia mensual'] # Si la persona recuerda o no será la variable respuesta
X_test = test_data.drop('frecuencia mensual', axis=1)
y_test = test_data['frecuencia mensual']
```

```python
y_test.head(3)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
```

```python
y_pred = lda.predict(X_test) # Se hace prediccion sobre los datos de validacion para la variable explicativa
print("Predicciones:", y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calcular matriz
cm = confusion_matrix(y_test, y_pred)

# Crear figura
plt.figure(figsize=(6,5))

# Dibujar heatmap bonito
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    cbar=True,
    linewidths=1,
    linecolor='white'
)

# Etiquetas y título
plt.title('Matriz de Confusión', fontsize=14, fontweight='bold')
plt.xlabel('Predicción')
plt.ylabel('Valor Real')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Matriz de Confusión - LDA")
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

modelo_rf = RandomForestClassifier(
    n_estimators=300,
    class_weight='balanced',
    random_state=42
)

modelo_rf.fit(X_train, y_train)

y_pred_rf = modelo_rf.predict(X_test)

print(classification_report(y_test, y_pred_rf))
```

```python

from sklearn.linear_model import LogisticRegression

modelo_log = LogisticRegression(
    max_iter=1000,
    class_weight='balanced'
)

modelo_log.fit(X_train, y_train)

y_pred_log = modelo_log.predict(X_test)

print(classification_report(y_test, y_pred_log))
```
