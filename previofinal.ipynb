```python
import re

# Excepción personalizada
class CorreoInvalidoError(Exception):
    pass

# Función de validación
def validar_correo(correo):

    patron = r"^[a-z]{3,12}+_\d+_[a-z]{3,30}+@universitaria\.edu\.co$"
    if not re.fullmatch(patron, correo):
        raise CorreoInvalidoError(
            "Correo inválido. Use el formato: nombre_edad_apellido@universitaria.edu.co"
        )
    return True

# Programa principal
INTENTOS = 3


while 0 < INTENTOS:
    try:
        correo = input("Ingrese su correo institucional: ")
        validar_correo(correo)
        print("Correo válido")
        break
    except CorreoInvalidoError as e:
        INTENTOS -= 1
        print("Error:", e)
        if 0 < INTENTOS:
            print(f"Intentos restantes: {INTENTOS}")
    except Exception as e:
        print("Error inesperado:", e)
        break

if INTENTOS == 0:
    print("Acceso bloqueado por intentos fallidos.")
```
