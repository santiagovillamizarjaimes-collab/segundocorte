```python
try:
  num1 = float(input("Número 1: "))
  num2 = float(input("Número 2: "))
  op = input("Operación (+, -, *, /, %): ")


  if op == "+":
    resultado = num1 + num2
  elif op == "-":
    resultado = num1 - num2
  elif op == "*":
    resultado = num1 * num2
  elif op == "/":
    resultado = num1 / num2
  elif op == "%":
    resultado = num1 % num2
  else:
    raise ValueError("Operación no válida")

  print("Resultado: ", resultado)

except ValueError as e:
  print("Error de valor: ", e)

except ZeroDivisionError as e:
  print("No se puede dividir entre cero")

except TypeError as e:
  print("Error de tipo: ", e)
```

```python
productos = {
    "pan": 1.5,
    "leche": 2.0,
    "huevos": 3.0
}

print("Productos disponibles:", productos)

try:
  nombre = input("¿Qué producto quieres?: ").lower()

  if nombre not in productos:
    raise KeyError("El producto no está disponible")

  cantidad = int(input("Cantidad: "))

  if cantidad <= 0:
    raise ValueError("La cantidad debe ser mayor a 0")

  total = productos[nombre] * cantidad
  print("Total a pagar:", total)

except KeyError as e:
  print("El producto no está disponible")
except ValueError as e:
  print("La cantidad debe ser un número entero positivo", e)
```

```python
import re
texto = "Tengo 23 perros y 3 gatos"
resultado = re.findall(r"\d", texto)
print(resultado)
```

el método findall permite encontrar números (\d significa dígitos). No determina cadenas de solo dígitos

```python
import re
texto = "Tengo 23 perros y 3 gatos"
resultado = re.findall(r"\d+", texto)
print(resultado)
```

```python
import re
texto = "Python es divertido"
resultado = re.findall(r"\w+", texto)
print(resultado)
```
