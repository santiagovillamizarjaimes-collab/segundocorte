Aproximadamente el 75% de las calificaciones son hasta 4.4

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, gaussian_kde

# Datos
data = df["nota"].dropna()

# Configuración estética
sns.set(style="whitegrid", context="talk")

# Crear figura
plt.figure(figsize=(10,6))

# Histograma (normalizado)
sns.histplot(data, bins=15, stat="density", color="skyblue", edgecolor="black", alpha=0.6, label="Histograma")

# Parámetros de la normal
mu, std = data.mean(), data.std()

# Rango de valores
x = np.linspace(data.min(), data.max(), 1000)

# FDP Normal
pdf_normal = norm.pdf(x, mu, std)
plt.plot(x, pdf_normal, color="red", linewidth=2.5, label=f"Normal ($\\mu$={mu:.2f}, $\\sigma$={std:.2f})")

# KDE (Kernel Density Estimation)
kde = gaussian_kde(data)
pdf_kde = kde(x)
plt.plot(x, pdf_kde, color="darkgreen", linewidth=2.5, linestyle="--", label="KDE (Kernel)")

# Detalles finales
plt.title("Distribución de la variable 'nota'", fontsize=16)
plt.xlabel("Nota")
plt.ylabel("Densidad")
plt.legend()
plt.tight_layout()

# Mostrar
plt.show()
```

Descriptivamente la gráfica bajo el Kernell se parece a la normal

```python
import numpy as np
import pandas as pd

np.random.seed(123)

grupo_A = np.random.normal(loc=4.0, scale=0.4, size=40)
grupo_B = np.random.normal(loc=4.3, scale=0.5, size=40)
```

```python
df = pd.DataFrame({
    "A": grupo_A,
    "B": grupo_B
})
```

```python
from scipy import stats
```

```python
t_stat, p_val = stats.ttest_ind(df["A"], df["B"])

print("Estadístico t:", t_stat)
print("Valor p:", p_val)
```

```python
%%writefile seguridad/registro.py
def pedir_usuario():
  return input("Ingrese su usuario: ")
def pedir_contrasena():
  return input("Ingrese su contraseña: ")
```

```python
%%writefile seguridad/validacion.py

def usuario_valido(usuario):


    if len(usuario) < 5:
        return False


    if " " in usuario:
        return False

    return True


def es_valida(contrasena, usuario):


    if len(contrasena) < 8:
        return False


    if contrasena == usuario:
        return False

    tiene_mayus = any(c.isupper() for c in contrasena)
    tiene_minus = any(c.islower() for c in contrasena)
    tiene_num = any(c.isdigit() for c in contrasena)
    tiene_esp = any(not c.isalnum() and not c.isspace() for c in contrasena)
    tiene_espacio = any(c.isspace() for c in contrasena)


    return tiene_mayus and tiene_minus and tiene_num and tiene_esp and not tiene_espacio
```

```python
%%writefile seguridad/mensaje.py

def mostrar_usuario_invalido():
    print("Usuario inválido")
    print("Debe tener mínimo 5 caracteres y no contener espacios")



def mostrar_intento(intento, maximo):
    print(f"\nIntento {intento} de {maximo}")


def mostrar_resultado(valida):

    if valida:
        print("Contraseña válida")
        print("Registro exitoso")

    else:
        print("Contraseña inválida")


def mostrar_bloqueo():
    print("\nSistema bloqueado")
    print("Ha superado el número máximo de intentos")
```
