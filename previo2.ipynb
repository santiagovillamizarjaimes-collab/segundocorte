```python
import os
os.makedirs('src', exist_ok=True)
```

```python
%%writefile src/productos.py

def pedir_producto():
  productos = []
  precios = []

  for i in range(3):
    nombre = input("Ingrese el nombre del producto: ")
    precio = int(input("Ingrese el precio del producto: "))

  return productos, precios
```

```python
%%writefile src/calculos.py

def calcular_total(precios):
  return sum(precios)

def calcular_iva(total):
  return total * 0.19
```

```python
%%writefile src/factura.py

def mostrar_factura(productos, precios, total, iva):
  print("---------Factura---------")

  for i in range(len(productos)):
    print(f"{productos[i]}: {precios[i]}")

  print("----------------------------")
  print("Total: ", total)
  print("IVA: ", iva)
  print("Total a pagar: ", total + iva)
```

```python
from src import productos
from src import calculos
from src import factura

producto_lista, precio_lista = productos.pedir_producto()

total = calculos.calcular_total(precio_lista)
iva = calculos.calcular_iva(total)

factura.mostrar_factura(producto_lista, precio_lista, total, iva)
```
