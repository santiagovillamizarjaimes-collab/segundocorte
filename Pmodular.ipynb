```python
suma = lambda a, b: a+b
suma(3,4)
```

```python
producto = lambda a, b: a*b
producto(2, 8)
```

```python
ultimoele = lambda palabra: palabra[-1]
ultimoele('Santiago')
```

```python
palindromo = lambda palabra: palabra[::-1]
palindromo("Hola mundo")
```

```python
lambda1 = list(map(lambda x: x**2, [5, 6, 7, 8, 9]))
lambda1
```

```python
par = lambda x: x % 2 == 0
par(0)
```

Filter es una función integrada en Python. Filtra valores de una lisata dada una condición boleana

```python
list(filter(par, [1, 2, 3, 4]))
```
